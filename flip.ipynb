# Flip Gambar

```python
# Untuk Meng-Import Library dari python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("1.jpg") # Mendeklarasikan gambar
```

```python
img_height = img.shape[0] # Untuk mendapat nilai tinggi gambar dari properti shape, angka 0 adalah indeks pertama dalam citra yg mewakili nilai tinggi
img_width = img.shape[1] # Untuk mendapat nilai lebar gambar dari properti shape, angka 1 adalah indeks kedua dalam citra yg mewakili nilai lebar
img_channel = img.shape[2] # untuk mendapatkan jumhlah channel dari properti shape, angka 2 adalah indeks ketiga dalam citra yg mewakili saluran warna
img_type = img.dtype # untuk mengambil tipe data gambar
```

```python
img_flip_horizontal = np.zeros(img.shape, img_type) #untuk membuat array baru pada bagian horizontal
img_flip_vertical = np.zeros(img.shape, img_type) #untuk membuat array baru pada bagian horizontal
```

```python
for y in range(0, img_height): # Looping variabel y
    for x in range(0, img_width): # Looping variabel x
        for c in range(0, img_channel): # Looping variabel c
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c] # untuk melakukan putaran horizontal pada gambar
```

```python
for y in range(0, img_height): # Looping variabel y
    for x in range(0, img_width): # Looping variabel x
        for c in range(0, img_channel): # Looping variabel c
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c] # untuk melakukan putataran vertical pada citra
```

```python
plt.imshow(img_flip_horizontal) # untuk menampilkan gambar hasil pemutaran horizontal dalam bentuk plot
plt.title("Flip Horizontal") # untuk menampilkan title pada gambar
plt.show() # menampilkan gambar hasil pemutaran horizontal
plt.imshow(img_flip_vertical)# untuk menampilkan gambar hasil pemutaran vertical dalam bentuk plot
plt.title("Flip Vertical") # untuk menampilkan title pada gambar
plt.show()# menampilkan gambar hasil pemutaran vertical
```
