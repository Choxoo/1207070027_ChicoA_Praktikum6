# Inversi, Transformasi Logaritmik, Transformasi Powerlaw

```python
# Untuk Meng-Import Library dari python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("1.jpg") # untuk mengdeklarasikan gambar
```

```python
img_height = img.shape[0] # Untuk mendapat nilai tinggi gambar dari properti shape, angka 0 adalah indeks pertama dalam citra yg mewakili nilai tinggi
img_width = img.shape[1] # Untuk mendapat nilai lebar gambar dari properti shape, angka 1 adalah indeks kedua dalam citra yg mewakili nilai lebar
img_channel = img.shape[2] # untuk mendapatkan jumhlah channel dari properti shape, angka 2 adalah indeks ketiga dalam citra yg mewakili saluran warna
```

# Inversi

```python
img_inversi = np.zeros(img.shape, dtype=np.uint8) # untuk membuat array kosong
```

```python
def inversi_grayscale(nilai): # deklarasi fungsi
    for y in range(0, img_height): # looping y
        for x in range(0, img_width): # looping x
            red = img[y][x][0]
            green = img[y][x][1] # untuk megambil nilai RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # untuk menghitung nilai rata-rata
            gray = nilai - gray
            img_inversi[y][x] = (gray, gray, gray) # inversi nilai gray baru yg sebelumnya sudah dihitung
```

```python
def inversi_rgb(nilai): # deklarasi fungsi
    for y in range(0, img_height): # Looping y
        for x in range(0, img_width): # Looping x
            red = img[y][x][0] # mengambil nilai komponen warna merah
            red = nilai - red # untuk melakukan pengurangan nilai merah dari nilai yg diberikan pada fungsi
            green = img[y][x][1]
            green = nilai - green # mengambil nilai komponen warna hijau
            blue = img[y][x][2]
            blue = nilai - blue # mengambil nilai komponen warna biru
            img_inversi[y][x] = (red, green, blue) # mengatur nilai warna pixel yg dihasilkan 
```

```python
inversi_grayscale(255) # memanggil fungsi dengan nilai 255
plt.imshow(img_inversi) # menampilkan gambar yg sudah di inversi
plt.title("Inversi Grayscale") # menampilkan title
plt.show() # menampilkan gambar dengan judul yg sudah ditentukan

inversi_rgb(255) # memanggil fungsi dengan nilai 255
plt.imshow(img_inversi) # menampilkan gambar yg sudah di inversi
plt.title("Inversi RGB") # menampilkan title
plt.show() # menampilkan gambar dengan judul yg sudah ditentukan
```

# Log


```python
img_log = np.zeros(img.shape, dtype=np.uint8) # membuat array kosong baru
```

```python
def log(c): # deklarasi funsi log dengan mengambil argumen c
    for y in range(0, img_height): # looping y
        for x in range(0, img_width): # looping x
            red = img[y][x][0]
            green = img[y][x][1] # untuk megambil nilai RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # untuk mengambil nilai rata" gray
            gray = int(c * np.log(gray + 1))
            if gray > 255: # nilai batas
                gray = 255 
            if gray < 0: # nilai batas
                gray = 0
            img_log[y][x] = (gray, gray, gray) # mengatur warna pixel yang dihasilkan
```

```python
log(30) # memanggil fungsi log dengan nilai 30
plt.imshow(img_log) # menampilkan gambar
plt.title("Log") # menampilkan title
plt.show() # menampilkan gambar dengan judul yg sudah ditentukan
```

# Inversi & Log

```python
img_inlog = np.zeros(img.shape, dtype=np.uint8) # membuat array kosong baru
```

```python
def inlog(c): # deklarasi funsi inlog dengan mengambil argumen c 
    for y in range(0, img_height): # looping x
        for x in range(0, img_width): # looping y
            red = img[y][x][0]
            green = img[y][x][1] # untuk megambil nilai RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # untuk mengambil nilai rata" gray
            gray = int(c * np.log(255 - gray + 1))
            if gray > 255: # nilai batas
                gray = 255
            if gray < 0: # nilai batas
                gray = 0
            img_inlog[y][x] = (gray, gray, gray) # mengatur warna pixel yang dihasilkan
```

```python
inlog(30) # memanggil fungsi inlog dengan nilai 30
plt.imshow(img_inlog) # menampilkan gambar
plt.title("Inversi & Log") # menampilkan title
plt.show() # menampilkan gambar dengan judul yg sudah ditentukan
```

# Nth Power

```python
img_nthpower = np.zeros(img.shape, dtype=np.uint8) # membuat array kosong baru
```

```python
def nthpower(c, y): # deklarasi fungsi dengan 2 argumen
    thc = c / 100 # mengubah nilai skala c dan y
    thy = y / 100
    for y in range(0, img_height): # looping y
        for x in range(0, img_width): # looping x
            red = img[y][x][0]
            green = img[y][x][1] # untuk megambil nilai RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # mengambil nilai rata" dari gray
            gray = int(thc * pow(gray, thy))
            if gray > 255: # nilai batas
                gray = 255
            if gray < 0: # nilait batas
                gray = 0
            img_nthpower[y][x] = (gray, gray, gray) # mengatur warna pixel yang dihasilkan
```

```python
nthpower(50, 100) # memanggil fungsi inlog dengan nilai 50, 100
plt.imshow(img_nthpower) # menampilkan gambar
plt.title("Nth Power") # menampilkan title
plt.show() # menampilkan gambar dengan judul yg sudah ditentukan
```

# Nth Root Power

```python
img_nthrootpower = np.zeros(img.shape, dtype=np.uint8) # membuat array kosong baru
```

```python
def nthrootpower(c, y): # deklarasi fungsi dengan 2 argumen
    thc = c / 100 # mengubah nilai skala c dan y
    thy = y / 100
    for y in range(0, img_height): # looping y
        for x in range(0, img_width): # looping x
            red = img[y][x][0]
            green = img[y][x][1] # untuk megambil nilai RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # mengambil nilai rata" dari gray
            gray = int(thc * pow(gray, 1./thy))
            if gray > 255: # nilai batas 
                gray = 255
            if gray < 0: # nilai batas
                gray = 0
            img_nthpower[y][x] = (gray, gray, gray) # mengatur warna pixel yang dihasilkan
```

```python
nthrootpower(50, 100) # memanggil fungsi inlog dengan nilai 50, 100
plt.imshow(img_nthrootpower) # menampilkan gambar
plt.title("Nth Root Power")
plt.show() # menampilkan gambar dengan judul yg sudah ditentukan
```
