# Histogram

```python
# Untuk Meng-Import Library dari python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("1.jpg") # # Mendeklarasikan gambar
```

```python
img_height = img.shape[0] # Untuk mendapat nilai tinggi gambar dari properti shape, angka 0 adalah indeks pertama dalam citra yg mewakili nilai tinggi
img_width = img.shape[1] # Untuk mendapat nilai lebar gambar dari properti shape, angka 1 adalah indeks kedua dalam citra yg mewakili nilai lebar
img_channel = img.shape[2] # untuk mendapatkan jumhlah channel dari properti shape, angka 2 adalah indeks ketiga dalam citra yg mewakili saluran warna
```

```python
img_grayscale = np.zeros(img.shape, dtype=np.uint8) # untuk membuat array baru pada grayscale

for y in range(0, img_height): # Looping variabel y
    for x in range(0, img_width): # Looping variabel x
        red = img[y][x][0]
        green = img[y][x][1] # untuk megambil nilai RGB
        blue = img[y][x][2]
        gray = (int(red) + int(green) + int(blue)) / 3 # untuk menghitung nilai rata-rata
        img_grayscale[y][x] = (gray, gray, gray) # megisi nilai RGB dengan hasil perhitungan rata" sebelumnya
        
plt.imshow(img_grayscale) # untuk menampilkan gambar dalam bentuk plot
plt.title("Grayscale") # untuk menampilkan title pada gambar
plt.show()# menampilkan hasil gambar
```

# Menampilkan Histogram Gambar Grayscale

```python
hg = np.zeros((256)) # membuat array kosong dengan ukuran 256
```

```python
for x in range(0, 256): # Looping untuk menjadikan semua nilai hg menjadi 0
    hg[x] = 0
```

```python
for y in range(0, img_height): # Looping y
    for x in range(0, img_width): # Looping x
        gray = img_grayscale[y][x][0] # untuk mengambil nilai pada grayscale
        hg[gray] += 1 # penambahan satu ke elemen array setiap nilai intensitas gray ditemukan
```

```python
# plt.figure(figsize=(20, 6))
# plt.plot(hg, color="black", linewidth=2.0)
# plt.show()

bins = np.linspace(0, 256, 100) # untuk menentukan batasan interval
plt.hist(hg, bins, color="black", alpha=0.5) # untuk menggambar histogram
plt.title("Histogram") # untuk menampilkan title
plt.show() # untuk menampilkan gambar histogram
```

# Menampilkan Histogram Kumulatif

```python
hgk = np.zeros((256)) # untuk membuat variabel hgk dengan array kosong dengan ukuran 256
c = np.zeros((256)) # untuk membuat variabel c dengan array kosong dengan ukuran 256

for x in range(0, 256):  # looping x dari 0 - 255
    hgk[x] = 0 # mengatur elemen x dalam variabel hgk menjadi 0
    c[x] = 0 # mengatur elemen c dalam variabel hgk menjadi 0

for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0] # untuk megambil nilai grayscale
        hgk[gray] += 1 # menambahkan 1 ke elemen array rgb
                
c[0] = hgk[0] # untuk menyimpan nilai 
for x in range(1, 256): # looping x dari 1 -255
     c[x] = c[x-1] + hgk[x] # menghitung nilai

hmaxk = c[255] # untuk menyimpan nilai variabel dari variabel c

for x in range(0, 256): # looping dari 0 - 255
    c[x] = 190 * c[x] / hmaxk

plt.hist(c, bins, color="black", alpha=0.5)  # menggambar histogram bewarna hitam dengan nilai transparansi 0.5
plt.title("Histogram Grayscale Kumulatif") # menampilkan title
plt.show() # menggambarkan histogram
```

# Menampilkan Histogram Hequalisasi

```python
hgh = np.zeros((256)) # untuk membuat variabel hgh dengan array kosong dengan ukuran 256
h = np.zeros((256)) # untuk membuat variabel h dengan array kosong dengan ukuran 256
c = np.zeros((256)) # untuk membuat variabel c dengan array kosong dengan ukuran 256

for x in range(0, 256): # looping x dari 0 - 255
    hgh[x] = 0 # mengatur elemen hgh dalam variabel hgk menjadi 0
    h[x] = 0 # mengatur elemen h dalam variabel hgk menjadi 0
    c[x] = 0 # mengatur elemen c dalam variabel hgk menjadi 0

for y in range(0, img_height): # Looping
    for x in range(0, img_width): # Looping
        gray = img_grayscale[y][x][0] # untuk megambil nilai grayscale
        hgh[gray] += 1 # menambahkan 1 ke elemen array rgb
                
h[0] = hgh[0] # untuk menyimpan nilai 
for x in range(1, 256): # looping x dari 1 -255
     h[x] = h[x-1] + hgh[x] # menghitung nilai


for x in range(0, 256): # looping x dari 0 -255
     h[x] = h[x] / img_height / img_width # Menghitung nilai

for x in range(0, 256): # looping x dari 0 -255
    hgh[x] = 0
    
for y in range(0, img_height): # Looping
    for x in range(0, img_width): # Looping
        gray = img_grayscale[y][x][0] # mengambil nilai abu"
        gray = h[gray] * 255  # mengalikan nilai varibel h
        hgh[int(gray)] += 1 # menambahkan 1 ke elemen array rgb

c[0] = hgh[0] # menyimpan nilai
for x in range(1, 256): # Looping x dari1 - 255
     c[x] = c[x-1] + hgh[x] # untuk melakukan perhitungan

hmaxk = c[255] # variabel yg menyimpan nilai variabel c 

for x in range(0, 256): # looping x 
    c[x] = 190 * c[x] / hmaxk # untuk melakukan perhitungan

plt.hist(c, bins, color="black", alpha=0.5)  # menggambar histogram bewarna hitam dengan nilai transparansi 0.5
plt.title("Histogram Grayscale Hequalisasi") # menampilkan title
plt.show() # menampilkan gambar histogram
```
